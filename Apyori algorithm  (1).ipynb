```python
for i in range (0,7051):
    transaction.append([str(data.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
rules= apriori(transactions=transaction,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_legth=2)
```

```python
results=list(rules)
print(results)
```

```python
def inspect(results):
    lhs=[tuple(result[2][0][0])[0]for result in results]
    rhs=[tuple(result[2][0][1])[0]for result in results] 
    Supports=[result[1]for result in results]
    Confidences=[result[2][0][2]for result in results]
    Lifts=[result[2][0][3]for result in results]
    return list(zip(lhs,rhs,Supports,Confidences,Lifts))

New_Table=pd.DataFrame(inspect(results),columns=['Left hand side','Right hand side','support','confidence','lift'])
```
